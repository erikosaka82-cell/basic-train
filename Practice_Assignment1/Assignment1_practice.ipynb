```python
# by E, Detete this later


# Classes and Function


# Step 1: create a class BankAccount
# Add class attribute bankname = "Canada Bank"
# Add 2 instance attributes: account_holder (string) and balance (floot, default value 0.0)

class BankAccount2:
  account_holder2: str # <--this is a code, and that is why using ':' , not ' = "" '
  balance2: float = 0.0 # <-- this is also a code.
  bankname2 = 'Canada Bank 2'


# Step 2: Add method

  def __init__(self, account_holder2: str, balance2 = 0.0) -> None:
      self.account_holder2 = account_holder2
      self.balance2 = balance2
      # ' -> None ' is a 'Type Hint', which means 'there is nothing to return = None' And used for clearfy the code writer's intension

  def deposit2(self, amount2:float) -> None:
      self.balance2 += amount2

#  " += " is an addition assignment operator.
# This is the same to " self.balance2 = self.balance2 + amount2 "
# add amount2 to current value "self.balance2 ", then return that sum of value to self.balance2

  def withdraw2(self, amount2:float) -> None:
      if self.balance2 >= amount2:
        self.balance2 -= amount2
      else:
        print("Insufficient funds. Please check your current balance")

  def get_balance2(self) -> None:
      name_balance = f"Current balance of \033[1m {self.account_holder2} \033[0m is \033[1m {self.balance2}\033[0m Canadian dollars"
      print(name_balance)

  @staticmethod
  def get_bankname2() -> None:
      print(bankname2)


# The difference between period and comma.
#  "self.name = name" can use ' .period '. But 'def speak(self, sound)' uses ' ,comma '
# ' .period ' is used to access attribute or method, means' .name ' is neccessary to access those info.
# On the other hand, ' ,comma ' is used to separate several arguments. both 'self' and 'sound' are arguments.




# Step3 Test
#  3.1, Print the Bank name
print(f" \033[1m Step3.1 \033[0m: print the bank name by using staticmethod ")

print(BankAccount2.bankname2)

print(f" \n")



# 3.2 Create an object of BankAccount2 with your name and an initial balance of $1000
print(holizontal_line)
print(f" \033[1m Step3.2 \033[0m: Create an object of BankAccount2 with your name and an initial balance of $1000 ")


client2 = BankAccount2("Ellie",1000)

print(f" Current Status \n Account Holder : {client2.account_holder2}, Current balance : {client2.balance2}")
print(f" \n")

# 3.3 Deposit 500
print(holizontal_line)
print(f" \033[1m Step3.3 \033[0m: Deposit 500 ")

client2.deposit2(500)

print(f" Current Status \n Account Holder : {client2.account_holder2}, Current balance : {client2.balance2}")
print(f" \n")


# 3.4 Withdraw 200
print(holizontal_line)
print(f" \033[1m Step3.4 \033[0m: Withdraw 200 ")

client2.withdraw2(200)

print(f" Current Status \n Account Holder : {client2.account_holder2}, Current balance : {client2.balance2}")
print(f" \n")


# 3.5 Print the final balance
print(holizontal_line)
print(f" \033[1m Step3.5 \033[0m: Print the final balance \033[1m by using get_balance() \033[0m ")


client2.get_balance2()
```

```python
# merging the Dataframe on 'studentID'
# pd.merge() requires first argument(->left), second argument(->right), and new frame in order to put this new merged frame.

# As a default, common row is used as a key when merging.
# ' on='key name' is used to indentify the key row ' --> print(pd.merge(df_abb, df_acc, on='')


# when there are some columns used both dataframe, "suffixes" is used to avoid overlapping (= two different columns are made)
# when there are some columns used both dataframe, "suffixes" is used to avoid making same name columns.
# In this case, by using " suffixied = ('_English', '_Math') ",  this suffix is used for each dataframe


df_studentGrades2 = pd.merge(df_english2, df_math2, on='studentID', suffixes=('_English', '_Math'))

print("This is the entire merged frame (= df_studentGrades2)")
print(f"\nStudnetID is chosen since this is a unique identifier in both csv files.")
print(f"StudentID does not have suffixes(= _English, or _Math) since StudentID is used as a key when merging both table \n \n ")
print(holizontal_line)
print(f" \n ")

print(df_studentGrades2)
```

```python
# calculate the average GPA across the two classes for each student


# .mean is used to output average.
# axis is used to indicate which direction is used for getting average.
# axis=0 means index order. axis=1 means columns order.

df_studentGrades2['GPA'] = df_studentGrades2[['grade_English','grade_Math' ]].mean(axis=1)

# df_studentGrade2['GPA] means adding new column indexed GPA
# = df_studentGrades[[]] is extracting certain columns to be caluculated.
# In dataframe, [[column1, column2,...]] can extract certain columns.

print(f"This is the dataframe with GPA column added")
print(holizontal_line)
print(f" \n")


print(df_studentGrades2.head(5))
```

```python
#  Delete duplicated columns (= rows for first name and last name)

df_studentGrades2=df_studentGrades2.drop(columns=[ 'firstName_Math', 'lastName_Math'])

```

```python
print(df_studentGrades2.head(5))
```

```python
# rename index and columns
# Dictionary method is used for renaming
# df_new = df_rename(columns={'Previous name of column1':'New name of column1', ,,,'Previous name of columnX:'New name of columnX'}, index = {'One':'Row1','''}

df_studentGrades2 = df_studentGrades2.rename(columns={
    'firstName_English':'First Name',
    'lastName_English':'Last Name',
    'grade_English':'English Grade',
    'grade_Math':'Math Grade'
    })



```

```python
# describe method
# in panda series,
# when data type is number, describe() returns count, mean, std, min, 25%, 50%,75%, max
# when data type is object(such as string), describe() returns  count, unique, top, freq


# count: the number of elements (Nan is not counted)
# mean: mathmatical average
# std : Standard deviation
# min :minimum
# 25% : 25 percentile
# 50% : 50 percentile = median
# 75% : 75 percentile
# max : maximum

print(df_studentGrades2.describe())
```

```python
# change a datatype for each columns by using astype

df_studentGrades2['studentID'] = df_studentGrades2['studentID'].astype(str)
df_studentGrades2['First Name'] = df_studentGrades2['First Name'].astype(str)
df_studentGrades2['Last Name'] = df_studentGrades2['Last Name'].astype(str)
df_studentGrades2['English Grade'] = df_studentGrades2['English Grade'].astype(float)
df_studentGrades2['Math Grade'] = df_studentGrades2['Math Grade'].astype(float)
df_studentGrades2['GPA'] = df_studentGrades2['GPA'].astype(float)

```

```python
# sort the merged dataframe so that it is in descending order based on GPA


# sort_values('name of index')  sorting by the values of each elemet. This is naturally ascending.
# '  ,ascending = false ' should be added to sort in descending order.


df_studentGrades2_GPAsorted = df_studentGrades2.sort_values('GPA', ascending = False)

print(df_studentGrades2_GPAsorted)
```

```python
# way of accessing the student with highest GPA

# .loc and .iloc are both methods of extracting values.
# .loc extract values by indicating NAME OF rows or columns (= label names).
# .iloc can extract values by indicating index orders.(= number of index order)


print(f"In this case, .iloc is preferable to select highest GPA student \nsince this method used index order to extract certain data.\nThis means highest data always comes to row 0. It is easy to find the highest score of data.")
print(f"\nOn the other hand, .loc requires the name of row and column to reach certain data. \nAfter sorting a dataframe, identify highest GPA students' name is laborious. ")
```

```python
# use .iloc assign the variable below to the row containging the student with the highest GPA


highest_student2 = df_studentGrades2_GPAsorted.iloc[0]

print(highest_student2)


```
